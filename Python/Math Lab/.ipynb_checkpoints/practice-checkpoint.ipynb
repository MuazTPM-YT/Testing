```python
# a) Find the angle between the curves 𝑟 = 4(1 + 𝑐𝑜𝑠𝑡) & 𝑟 = 5(1 − 𝑐𝑜𝑠𝑡)

from sympy import *
import math

t = Symbol('t')

r1 = 4*(1 + 𝑐𝑜𝑠(𝑡))
r2 = 5*(1 - 𝑐𝑜𝑠(𝑡))

dr1 = diff(r1,t)
dr2 = diff(r2,t)

tanphi1 = r1/dr1
tanphi2 = r2/dr2

q = solve(r1-r2,t)

w1 = tanphi1.subs({t: q[1]})
w2 = tanphi2.subs({t: q[1]})

phi1 = atan(w1)
phi2 = atan(w2)

angle = abs(phi1-phi2)

print(f"{angle:.3f}")
print(math.degrees(angle))
```

```python
# b) Find the angle between the curves 𝑟 = 4 𝑐𝑜𝑠𝑡 & 𝑟 = 5 𝑠𝑖𝑛𝑡.

from sympy import *
import math

t = Symbol('t')

r1 = 4*cos(t)
r2 = 5*sin(t)

dr1 = diff(r1,t)
dr2 = diff(r2,t)

tanphi1 = r1/dr1
tanphi2 = r2/dr2

q = solve(r1-r2,t)

w1 = tanphi1.subs({t:q[0]})
w2 = tanphi2.subs({t:q[0]})

phi1 = atan(w1)
phi2 = atan(w2)

angle = abs(phi1-phi2)

print(f"Radian Angle: {angle:.3f}")
print(f"Degree Angle: {math.degrees(angle)}")
```

```python
1# c) Find the radius of curvature for the curve 𝑟 = 4(1 + 𝑐𝑜𝑠𝑡) 𝑎𝑡 𝑡 = pi/2

from sympy import *

t = Symbol('t')
r = 4*(1+cos(t))
r1 = diff(r,t)
r2 = diff(r1,t)

roc = ((r**2 + r1**2)**1.5)/(r**2 + 2*r1**2 - r2*r)
roc1 = roc.subs({t: pi/2})

display(simplify(roc))
print(roc1)
```

```python
# If u = xy/z, v = yz/x, w= zx/y then prove that J = 4.

from sympy import *
import numpy as np

init_printing()

x,y,z = symbols('x,y,z')

u = x*y/z
v = y*z/x
w = z*x/y

ux = diff(u,x)
uy = diff(u,y)
uz = diff(u,z)

vx = diff(v,x)
vy = diff(v,y)
vz = diff(v,z)

wx = diff(w,x)
wy = diff(w,y)
wz = diff(w,z)

A = Matrix([[ux,uy,uz],[vx,vy,vz],[wx,wy,wz]])
J = det(A)

display(A)
print(J)
```

```python
# a)Write a Program to find the Maclaurin’s series expansion of \
# sin(𝑥) + cos (𝑥) up to 3rd degree term.

from sympy import *

x = Symbol('x')
y = sin(x) + cos(x)
dy1 = diff(y,x)
dy2 = diff(dy1,x)
dy3 = diff(dy2,x)

y = lambdify(x,y)
y1 = lambdify(x,dy1)
y2 = lambdify(x,dy2)
y3 = lambdify(x,dy3)

y = y(0) + x*y1(0) + (x**2/2)*y2(0) + (x**3/6)*y3(0)
display(y)
```

```python
def gcd(a,b):
    if a<b:
        a,b=b,a
    r = a%b
    if r==0:
        return b
    else:
        return gcd(b,r)

print(gcd(2,8))
```

```python
# Write a Program to find the solution of the congruence 6𝑥 ≡ 5(𝑚𝑜𝑑 15).

a = 5
b = 3
n = 13

d = gcd(a,n)
r = b%d

if r!=0:
    print("No")
else:
    for k in range(1,n-1):
        x = (n*k + b)/a
        if x//1 == x:
            print(x)
            break
```

```python
import numpy as np

A = np.matrix([[1,2,-1],[2,1,4],[3,3,4]])
B = np.matrix([[1],[2],[1]])

rA = np.linalg.matrix_rank(A)

AB = np.concatenate((A,B),axis=1)
rAB = np.linalg.matrix_rank(AB)
n = A.shape[1]

if rA == rAB:
    if rA==n:
        print("Unique")
        print(np.linalg.solve(A,B))
    else:
        print("Infinite")
else:
    print("No Solution")
```
