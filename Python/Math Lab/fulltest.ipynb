```python
# 1) Find the angle between the curves r = a(1 + cos theta) and r = b(1 − cos
#    theta) where a=4 and b=5

from sympy import *
import math

theta = Symbol('theta')
a = 4
b = 5

r1 = a + a*cos(theta)
r2 = b - b*cos(theta)

dr1 = diff(r1, theta)
dr2 = diff(r2, theta)

tanphi1 = r1/dr1
tanphi2 = r2/dr2

q = solve(r1-r2, theta)

w1 = tanphi1.subs({theta: q[1]})
w2 = tanphi2.subs({theta: q[1]})

phi1 = atan(w1)
phi2 = atan(w2)

angle = abs(phi1 - phi2)

print(f"Angle Between the Two Curves (in Radians): {angle:.4f} radians")
print(f"Angle Between the Two Curves (in Degrees): {math.degrees(angle):.4f} degrees")
```

```python
# 2) Find the angle between the curves r = a log(theta) and r= a/log(theta) where a=4

from sympy import *
import math

theta = Symbol('theta')
a = 4

r1 = a*log(theta)
r2 = a/log(theta)

dr1 = diff(r1,theta)
dr2 = diff(r2,theta)

tanphi1 = r1/dr1
tanphi2 = r2/dr2

q = solve(r1-r2, theta)

w1 = tanphi1.subs({theta: q[1]})
w2 = tanphi2.subs({theta: q[1]})

phi1 = atan(w1)
phi2 = atan(w2)

angle = abs(phi1-phi2)

print(f"Angle between the curves (in degrees): {math.degrees(angle):.4f}")
```

```python
# 3) Find the radius of curvature, r = 4(1 + cos t) at t=π/2

from sympy import *

init_printing()
t = Symbol('t')

r = 4 + 4*cos(t)
r1 = diff(r,t)
r2 = diff(r1,t)

roc = ((r**2 + r1**2)**1.5) / (r**2 + 2*r1**2 - r2*r)
roc1 = roc.subs({t: pi/2})

print(f"ROC: {roc1:.4f} units")
display(simplify(roc))
```

```python
# 4) Find the radius of curvature for r = asin(nt) at t = pi/2 and n = 1

from sympy import *

a,n,t = symbols('a,n,t')
r = a*sin(n*t)

r1 = diff(r, t)
r2 = diff(r1, t)

roc = ((r**2 + r1**2)**1.5)/(r**2 + 2*r1**2 - r2*r)
roc1 = roc.subs({t: pi/2, n: 1})

print(f"ROC:")
display(roc1)
```

```python
# 1)  Prove that mixed partial derivatives, uxy= uyx for
#     u = exp(x)(xcos(y)-ysin(y)).

from sympy import *

x,y = symbols('x,y')
u = exp(x)*(x*cos(y)-y*sin(y))

ux = diff(u,x)
uy = diff(u,y)

uxy = diff(ux,y)
uyx = diff(uy,x)

if uxy==uyx:
    print(f"Mixed Partial Derivatives are Equal")
else:
    print(f"Mixed Partial Derivatives are not Equal")
```

```python
# 2) Prove that if u = exp(x)(x cos(y) − y sin(y)) then uxx+ uyy= 0.

from sympy import *

u = exp(x)*(x*cos(y)-y*sin(y))

ux = diff(u,x)
uy = diff(u,y)

uxx = diff(ux,x)
uyy = diff(uy,y)

w = uxx+uyy

print(f"Ans: {simplify(w)}")
```

```python
# 3) If u = xy/z, v = yz/x,w= zx/y then prove that J = 4.

from sympy import *

x,y,z = symbols('x,y,z')

u = (x*y)/z
v = (y*z)/x
w = (z*x)/y

ux = diff(u,x)
uy = diff(u,y)
uz = diff(u,z)

vx = diff(v,x)
vy = diff(v,y)
vz = diff(v,z)

wx = diff(w,x)
wy = diff(w,y)
wz = diff(w,z)

A = Matrix([[ux,uy,uz],[vx,vy,vz],[wx,wy,wz]])
J = det(A)

print(f"Jacobian of the Matrix A: {J}")
```

```python
def gcd(a,b):
    if a<b:
        a,b = b,a
    r=a%b
    if r==0:
        return b
    else:
        return gcd(b,r)

print(f"GCD: {gcd(33,44)}")
```
